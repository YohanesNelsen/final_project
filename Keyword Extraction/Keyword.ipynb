```python
urls = [
    'https://www.investopedia.com/terms-beginning-with-a-4769351', 
    'https://www.investopedia.com/terms-beginning-with-b-4769352',  
]


all_terms = []

for url in urls:
    options = ChromeOptions()
    options.use_chromium = True 

    driver = Chrome(options=options)

    driver.maximize_window()
    driver.get(url)

    elements = driver.find_elements(By.CSS_SELECTOR, ".dictionary-top300-list__list.mntl-text-link")

    terms = [element.text for element in elements]
    all_terms.extend(terms)

    print(f"Terms collected from {url}: {terms}")


print(f"All collected terms: {all_terms}")
driver.quit()
```

# EXTRACT ALL

```python
# URL to scrape
url = 'https://www.investopedia.com/financial-term-dictionary-4769738'

options = ChromeOptions()
options.use_chromium = True 

# Initialize Edge WebDriver with options
driver = Chrome(options=options)

driver.maximize_window()
driver.get(url)

# Wait for the page to load (optional, consider using WebDriverWait for better reliability)
driver.implicitly_wait(10)

# Dictionary to store terms for each letter
dictionary_terms = {}

# Extract terms for each subsection by identifying the unique class names for links
terms_elements = driver.find_elements(By.CSS_SELECTOR, "div.comp.dictionary-top24-list__sublist-content.mntl-block a")
for term in terms_elements:
    # Extract text from each link element (term name)
    term_text = term.text.strip()
    # Determine which letter section it belongs to based on the URL
    letter = term.get_attribute('href').split('/')[-1][0]
    if letter.isdigit():
        letter = '#'
    if letter not in dictionary_terms:
        dictionary_terms[letter] = []
    dictionary_terms[letter].append(term_text)

# Close the browser
driver.quit()

# Print all collected terms
print("Collected terms from A to Z:")
for letter, terms in dictionary_terms.items():
    print(f"{letter.upper()}: {terms}")
```
